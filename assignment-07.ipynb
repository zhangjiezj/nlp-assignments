```python
data = data[data['content'].notnull()]
data_xinhua_pos = data[data['source'].fillna('').str.contains('新华')]
data_xinhua_neg = data[data['source'].fillna('').str.contains('新华')==0]
print(len(data_xinhua_pos))
print(len(data_xinhua_neg))
print(len(data_xinhua_pos)/len(data))
```

```python
N=4000
data_xinhua_neg_sample = data_xinhua_neg.sample(N)
data_xinhua_pos_sample = data_xinhua_pos.sample(N)
print(len(data_xinhua_pos_sample))
```

```python
data_samples = data_xinhua_neg_sample.append(data_xinhua_pos_sample)
print(len(data_samples))
data_samples['content'][:10]
```

```python
import jieba
import re
#attributes=['author','source','content','feature','title']
attributes=['content']
def cut(str):
    return ' '.join(jieba.cut(re.sub(r'[^\w\s]',' ',str.strip())))
def propropess(sourcedata):
    for att in attributes:
        sourcedata[att]=sourcedata[att].fillna('').apply(cut)
    return sourcedata[att],sourcedata['is_xinhua']

x_samples,y_samples = propropess(data_samples)

                                    
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x_tfidf_samples = vectorizer.fit_transform(x_samples).toarray()
```

### Step5: 在traning_data, validation_data, test_data 上观察其相关metric: recall, precision, f1等， 并解释其含义. 

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
def get_performance(clf,x_,y_):
    y_hat = clf.predict(x_)
    f1score = f1_score(y_,y_hat)
    accuracyscore=accuracy_score(y_,y_hat)
    precisionscore=precision_score(y_, y_hat)
    recallscore=recall_score(y_, y_hat)
    print("f1_score is ",f1score)
    print("accuracy_score is ",accuracyscore)
    print("precision_score is ",precisionscore)
    print("recall_score is ",recallscore)
```

```python
get_performance(neighs,x_test,y_test)
```
